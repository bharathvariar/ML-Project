# Task 1

```python
#returns a new modified dataframe and does not change the input dataframe.
def feature1(data):
    data1 = data.copy()
    mean = np.mean(data1, axis=0)
    for i in range(2,data1.shape[1]):
        data1.iloc[:,i].fillna(mean[i-1], inplace=True)
    return data1
```

```python
#does not change the input dataframe.
#input the training dataframe without dropping any columns.
#returns X data, Y labels, mean, standard deviation.
#you can use mean and stddev to normalize test data if you have to, otherwise remove it from return.
def feature2(data):
    data1 = data.copy()
    y = data1['diagnosis']
    x = data1.drop(['diagnosis'], axis=1)
    mean = np.mean(x, axis=0)
    stddev = np.std(x, axis=0)
    x = (x - mean) / stddev
    return x, y, mean, stddev
```

```python
df2, y, mean, stddev = feature2(df)
```
